# scikit-learn

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier()
```

```python
data.head()
```

```python
data['sex'] = data['sex'] == 'Female'
data['smoker'] = data['smoker'] == 'Yes'
data['time'] = data['time'] == 'Dinner'
```

```python
data.head()
```

```python
data['good_tip'] = data['perc_tip'] > data['perc_tip'].mean()
```

```python
x = data.drop(['good_tip', 'day', 'perc_tip', 'tip'], axis=1)
y = data['good_tip']
```

```python
data.head()
```

```python
res = clf.fit(x, y)
```

```python
# STO BARANDO! STO BARANDO!
res.score(x, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
tree = DecisionTreeClassifier(max_depth=4)
```

```python
res = tree.fit(x, y)
```

```python
res.score(x, y)
```

```python
dot_data = export_graphviz(tree, out_file=None,
                         feature_names=x.columns,
                         filled=True, rounded=True,
                         special_characters=True)
```

```python
import graphviz
graph = graphviz.Source(dot_data)  
```

```python
graph
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(max_depth=4)
```

```python
res = forest.fit(x, y)
```

```python
res.score(x, y)
```
